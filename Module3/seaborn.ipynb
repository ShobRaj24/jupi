#Basic plotting using seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#%matplotlib inline
```

```python
mtcars=pd.read_csv('/content/mtcars.csv')
```

```python
mtcars.head()
```

```python
mtcars.info()
```

```python
mtcars.shape
```

##Barplot

```python
res=sns.barplot(x='cyl',y='carb', data=mtcars, color='yellow')
plt.show()
```

```python
res=sns.barplot(x='cyl',y='carb', data=mtcars, palette='rocket')
plt.show()
```

##Countplot

```python
#countplot returns the count of total values for each category using bars.
#number of vehicles for each category of cylinders
sns.countplot(x='cyl',data=mtcars,palette="Set1")
```

```python
#display the number of cars for each carburetor
sns.countplot(x='carb',data=mtcars,palette="Set1")
```

```python
sns.countplot(y='gear',data=mtcars,palette='rocket')
```

```python
#grouped countplot using hue parameter.
#displays the count of cars for each category of gears
#that are grouped based on the number of cylinders.
sns.countplot(x='gear',hue='cyl',data=mtcars,palette='Set1')
```

##Distribution plot

```python
#Distribution plot creates the distribution of any continuous data.
#distribution of mpg of the different vehicles.
sns.distplot(mtcars.mpg, bins=10, color='g')
```

##Scatterplot
##helps to create plots that can draw relationships between two continuous variables.

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
sns.scatterplot(x='sepal_length',y='petal_length',data=iris, hue='species')
```

##Pairplot
##The Python Seaborn library lets you visualize data using pair plots that produce a matrix of relationships between each variable in the dataset.

```python
#In the below plot, all the plots are histograms that represent
#the distribution of each feature.
sns.pairplot(iris)
```

```python
#KDE plots- Kernel Density estimate plots
sns.pairplot(iris, hue='species', palette='Set1')
```

##Boxplot
##A boxplot, also known as a box and whisker plot, depicts the distribution of quantitative data. The box represents the quartiles of the dataset. The whiskers show the rest of the distribution, except for the outlier points.

```python
sns.boxplot(x='species', y='sepal_width', data=iris)
```

##Linear regression plot
###The lmplot() function in the Seaborn library draws a linear relationship as determined through regression for the continuous variables.

```python
sns.lmplot(x='petal_length',y='petal_width',data=iris)
```

```python
sns.lmplot(x='petal_length',y='petal_width',hue='species',data=iris,markers=['o','*','^'])
```

##Heatmap

```python
import seaborn as sns
glue = sns.load_dataset("glue").pivot(index="Model", columns="Task", values="Score")
glue.head()
```

```python
sns.heatmap(glue)
```

```python
sns.heatmap(glue, annot=True)
```

```python
sns.heatmap(glue, annot=True, fmt=".1f")
```

```python
sns.heatmap(glue, annot=glue.rank(axis="columns"))
```

```python
sns.heatmap(glue, annot=True, linewidth=.5)
```

```python
sns.heatmap(glue, cmap="viridis") #crest, inferno, viridis, plasma
```

```python
sns.heatmap(glue, vmin=50, vmax=100)
```

```python
#using methods on matplotlib.axes.Axes object to tweak the plot
ax = sns.heatmap(glue, annot=True)
ax.set(xlabel="", ylabel="")
ax.xaxis.tick_top()
```
